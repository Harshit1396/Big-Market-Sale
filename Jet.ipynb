```python
skill['Premiumvalold'] = skill['Premiumvalold'].astype('int')
skill['Premiumval'] = skill['Premiumval'].astype('int')
```

```python
skill['Premiumvalold'] = skill['Skill Premium Old'].str.extract('(\d+)')
skill['Premiumval'] = skill['Skill Premium'].str.extract('(\d+)')
```

```python
weight_avg = skill.groupby('Emp Name').Premiumvalold.mean()
miss_bool = (skill['Premiumvalold'] != 1) 
def getweight(my):
    return weight_avg[my]
skill.loc[miss_bool,'Avgpremiumold'] = skill.loc[miss_bool,'Emp Name'].apply(getweight)
```

```python
weight_max = skill.groupby('Emp Name').Premiumvalold.max()
miss_bool = (skill['Premiumvalold'] != 1) 
def getweight(my):
    return weight_max[my]
skill.loc[miss_bool,'maxpremiumold'] = skill.loc[miss_bool,'Emp Name'].apply(getweight)
```
